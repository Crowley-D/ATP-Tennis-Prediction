# Pre Clean Datasets from Jeff Sackmann Datasets

Imports

```python
%matplotlib inline
import os
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
from collections import defaultdict
import re
import warnings
```

Import Datasets

```python
jeff_data = pd.DataFrame()
for year in range(1991, 2025):
    file = "./data/JeffSackmann/atp_matches_"+str(year)+".csv"
    
    year_data = pd.read_csv(file)

    jeff_data = pd.concat([jeff_data, year_data], axis=0)
    
jeff_data['tourney_date'] = pd.to_datetime(jeff_data['tourney_date'], format='%Y%m%d')

print(len(jeff_data))
jeff_data.head(5)
```

### Cleanup Rows

```python
jeff_data = jeff_data.dropna(subset=['winner_id', 'winner_name'], how='all')
jeff_data = jeff_data.dropna(subset=['loser_id', 'loser_name'], how='all')
jeff_data = jeff_data.dropna(subset=['tourney_id', 'tourney_name'], how='all')
jeff_data = jeff_data.dropna(subset=['round', 'tourney_date'], how='any')
```

### Cleanup Columns
1. Drop Unnecessary
2. Rename

```python
jeff_data = jeff_data.drop(columns=['winner_ioc', 'loser_ioc', 'winner_entry', 'loser_entry', 'winner_seed', 'loser_seed'], axis=1)
jeff_data = jeff_data.rename(columns={
    'winner_rank_points': 'winner_atp_points',
    'loser_rank_points': 'loser_atp_points',})

jeff_data.head(5)
```

### Tournament ID formatting

1. Davis Cup Standardisation

```python
def update_davis_cup_tourney_ids(df):
    """
    Update tourney_id for Davis Cup entries based on the word following 'Davis Cup'
    Finals: 430, G1/WG1/WG: 431, WG2/G2: 432, QLS: 433
    """
    # Create a copy to avoid modifying original
    df_updated = df.copy()
    
    # Extract Davis Cup entries
    davis_cup_mask = df_updated['tourney_name'].str.contains('Davis Cup', na=False, case=False)
    
    # Extract the word following "Davis Cup"
    pattern = r'Davis Cup\s+(\w+)'
    words_after = df_updated.loc[davis_cup_mask, 'tourney_name'].str.extract(pattern, flags=re.IGNORECASE)[0]
    
    # Create mapping based on the word
    for idx, word in words_after.items():
        if pd.isna(word):
            continue
        
        if word.upper() == 'FINALS':
            new_id = "430"
            new_name = "Davis Cup Finals"
        elif word.upper() in ['G1', 'WG1', 'WG']:
            new_id = "431"
            new_name = "Davis Cup World Group 1"
        elif word.upper() in ['WG2', 'G2']:
            new_id = "432"
            new_name = "Davis Cup World Group 2"
        elif word.upper() == 'QLS':
            new_id = "433"
            new_name = "Davis Cup Qualifiers"
        else:
            continue  # Keep original ID if no match
            
        df_updated.loc[idx, 'tourney_id'] = new_id
        df_updated.loc[idx, 'tourney_name'] = new_name
        df_updated.loc[idx, 'tourney_level'] = 'D'
    
    return df_updated

jeff_data = update_davis_cup_tourney_ids(jeff_data)
```

2. Olympics Standardisation

```python
olympic_mask = jeff_data['tourney_name'].str.contains('Olympic', na=False, case=False)
jeff_data.loc[olympic_mask, 'tourney_level'] = 'O'
jeff_data.loc[olympic_mask, 'tourney_id'] = 4800
```

2. Dropping Year Prefix

```python
def drop_year_prefix(df):
    """
    Drop everything up to and including the hyphen in tourney_id, only if hyphen exists
    """
    df_updated = df.copy()
    
    # Only split on hyphen if it exists
    mask_with_hyphen = df_updated['tourney_id'].str.contains('-', na=False)
    df_updated.loc[mask_with_hyphen, 'tourney_id'] = df_updated.loc[mask_with_hyphen, 'tourney_id'].str.split('-').str[1]
    
    return df_updated

# Apply the function
jeff_data = drop_year_prefix(jeff_data)
```

3. Remove Letters

```python
def clean_tourney_id(tourney_id):
    """
    Remove any letters from tourney_id and replace with '0'.
    Keeps only digits, hyphens, and underscores.
    
    Examples:
        '2024-580' -> '2024-580'
        '580' -> '580'
        '580A' -> '5800'
        'WTA-2024' -> '0-2024'
        'ITF123M' -> '0012300'
    """
    if pd.isna(tourney_id):
        return tourney_id
    
    tourney_id_str = str(tourney_id)
    
    # Replace any letter (a-z, A-Z) with '0'
    cleaned = re.sub(r'[a-zA-Z]', '0', tourney_id_str)
    
    return cleaned

# Apply cleaning to tourney_id column
print("Before cleaning:")
print(f"  Total rows: {len(jeff_data)}")
print(f"  Null tourney_id: {jeff_data['tourney_id'].isna().sum()}")
print(f"  Sample tourney_ids: {jeff_data['tourney_id'].unique()[:20]}")

# Check which tourney_ids contain letters
ids_with_letters = jeff_data[jeff_data['tourney_id'].astype(str).str.contains(r'[a-zA-Z]', na=False)]
print(f"\nRows with letters in tourney_id: {len(ids_with_letters)}")
if len(ids_with_letters) > 0:
    print("Sample IDs with letters:")
    print(ids_with_letters[['tourney_id', 'tourney_name']].drop_duplicates().head(20))

# Apply cleaning
jeff_data['tourney_id'] = jeff_data['tourney_id'].apply(clean_tourney_id)

print("\nAfter cleaning:")
print(f"  Null tourney_id: {jeff_data['tourney_id'].isna().sum()}")
print(f"  Sample tourney_ids: {jeff_data['tourney_id'].unique()[:20]}")

# Verify no letters remain
ids_with_letters_after = jeff_data[jeff_data['tourney_id'].astype(str).str.contains(r'[a-zA-Z]', na=False)]
print(f"\nRows with letters after cleaning: {len(ids_with_letters_after)}")

if len(ids_with_letters_after) == 0:
    print("✓ All letters successfully removed from tourney_id!")
```

4. Ensure 4 Digit Length

```python
jeff_data.copy().assign(
    tourney_id=jeff_data['tourney_id'].apply(
        lambda x: str(x) + '0' * (4 - len(str(x))) if pd.notna(x) and len(str(x)) < 4 else x
    )
)
```

4. Check Format

```python
def check_tourney_id_format(df):
    """
    Check for tourney_id values with hyphens, non-alphanumeric characters, or longer than 5 characters
    """
    issues_list = []
    
    for idx, row in df.iterrows():
        tourney_id = row['tourney_id']
        tourney_name = row['tourney_name']
        
        if pd.isna(tourney_id):
            issues_list.append([idx, tourney_id, tourney_name, "Missing/NaN value"])
            continue
            
        tourney_id_str = str(tourney_id)
        
        # Check for hyphens
        if '-' in tourney_id_str:
            issues_list.append([idx, tourney_id, tourney_name, "Contains hyphen"])
            continue
            
        # Check for non-alphanumeric characters
        if not tourney_id_str.isalnum():
            issues_list.append([idx, tourney_id, tourney_name, "Contains non-alphanumeric characters"])
            continue
            
        # Check length
        if len(tourney_id_str) > 5:
            issues_list.append([idx, tourney_id, tourney_name, "Longer than 5 characters"])
            continue
    
    # Convert to DataFrame
    if issues_list:
        format_issues_df = pd.DataFrame(issues_list, columns=['row_index', 'tourney_id', 'tourney_name', 'issue'])
    else:
        format_issues_df = pd.DataFrame(columns=['row_index', 'tourney_id', 'tourney_name', 'issue'])
    
    return format_issues_df

# Check format issues
format_issues = check_tourney_id_format(jeff_data)

print(f"Total format issues found: {len(format_issues)}")
```

```python
def check_tourney_id_compatibility(df):
    """
    Check if tourney_id values are compatible with Main_Matches_Dataset_Pipeline validation.
    The pipeline tries to convert to numeric with errors='coerce', which creates nulls for non-numeric values.
    """
    issues_list = []
    
    for idx, row in df.iterrows():
        tourney_id = row['tourney_id']
        tourney_name = row['tourney_name']
        
        if pd.isna(tourney_id):
            issues_list.append([idx, tourney_id, tourney_name, "Missing/NaN value"])
            continue
        
        tourney_id_str = str(tourney_id)
        
        # Check if it's purely numeric (can be converted to int/float without loss)
        try:
            numeric_value = pd.to_numeric(tourney_id_str, errors='coerce')
            if pd.isna(numeric_value):
                issues_list.append([idx, tourney_id_str, tourney_name, "Cannot be converted to numeric (would become NULL in pipeline)"])
        except:
            issues_list.append([idx, tourney_id_str, tourney_name, "Conversion error"])
    
    if issues_list:
        compatibility_issues_df = pd.DataFrame(issues_list, columns=['row_index', 'tourney_id', 'tourney_name', 'issue'])
    else:
        compatibility_issues_df = pd.DataFrame(columns=['row_index', 'tourney_id', 'tourney_name', 'issue'])
    
    return compatibility_issues_df

# Check compatibility
compatibility_issues = check_tourney_id_compatibility(jeff_data)

print(f"Total compatibility issues found: {len(compatibility_issues)}")
if len(compatibility_issues) > 0:
    print("\nSample issues:")
    print(compatibility_issues.head(20))
    
    # Show unique tourney_ids that have issues
    print(f"\nUnique problematic tourney_ids: {compatibility_issues['tourney_id'].nunique()}")
    print(compatibility_issues[['tourney_id', 'tourney_name']].drop_duplicates())
```

### Reformat Columns

```python
def rename_columns(df):
    col_map = {}
    for col in df.columns:
        new_col = col
        # Replace 'winner' with 'player1', 'loser' with 'player2'
        new_col = new_col.replace('winner', 'player1').replace('loser', 'player2')
        # Replace 'w_' with 'p1_', 'l_' with 'p2_'
        if new_col.startswith('w_'):
            new_col = new_col.replace('w_', 'p1_')
        if new_col.startswith('l_'):
            new_col = new_col.replace('l_', 'p2_')
        col_map[col] = new_col
    return df.rename(columns=col_map)
jeff_data = rename_columns(jeff_data)

jeff_data['source'] = 0
jeff_data = jeff_data.reindex(columns=['tourney_name', 'tourney_id', 'tourney_date', 'tourney_level', 'surface', 'draw_size', 'match_num', 'round', 'best_of', 'minutes', 'player1_id', 'player1_name', 'player2_id', 'player2_name', 'player1_rank', 'player1_atp_points', 'player2_rank', 'player2_atp_points', 'player1_ht', 'player1_hand', 'player2_ht', 'player2_hand', 'p1_ace', 'p2_ace', 'p1_df', 'p2_df', 'p1_SvGms', 'p2_SvGms', 'p1_1stIn', 'p2_1stIn', 'p1_1stWon', 'p2_1stWon', 'p1_2ndWon', 'p2_2ndWon', 'p1_svpt', 'p2_svpt', 'p1_bpSaved', 'p2_bpSaved', 'p1_bpFaced', 'p2_bpFaced', 'score'], copy=False)

jeff_data.tail()
```
